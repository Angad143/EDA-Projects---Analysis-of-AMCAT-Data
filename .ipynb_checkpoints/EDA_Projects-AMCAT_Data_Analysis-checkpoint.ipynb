## **EDA Project - AMCAT Data Analysis**

### **Step - 01 ---> Introduction**

The AMCAT (Advanced Mathematics and Computing Examination) dataset contains information about students' performance in various subjects.The dataset is derived from the Aspiring Mind Employment Outcome 2015 (AMEO) study, focusing on employment outcomes among engineering graduates. It includes information on various aspects such as salary, job titles, job locations, standardized scores in cognitive, technical, and personality skills, as well as demographic features. The dataset contains approximately 4000 data points with around 40 independent variables, comprising both continuous and categorical data.

#### **Objectives:** 
The objective of this exploratory data analysis (EDA) is to gain insights into the factors influencing the salary of engineering graduates. By examining the relationships between different variables and the target variable (Salary), we aim to understand the dataset's characteristics, identify patterns, and uncover potential insights that could aid in decision-making processes related to employment and salary negotiation for engineering graduates.

The objective of this analysis is to explore the dataset, perform univariate and bivariate analysis.


- 1.Explore employment outcomes, focusing on salary, job titles, and locations.

- 2.Understand data distribution and identify outliers.

- 3.Investigate relationships between variables.

- 4.Address research questions regarding salary and gender-specialization preferences.

- 5.Provide actionable insights for career development, recruitment, and education in the engineering domain.

### **I) For Numerical Columns:**

- #### **Understand the probability (PDF) and frequency distribution (Histogram) of each numerical column**
        - Histogram: Discrete, frequency-based representation.
        - PDF: Continuous, probability-based representation.

              -------> Both Histograms and PDFs provide visualizations of the frequency distribution of numerical data.

```python
# Histogram and PDFs of GraduationYear
plt.figure(figsize=(8, 5))
plt.xticks(rotation=90)

# Plot Histogram
sns.histplot(data["GraduationYear"], color="red", kde=False, label="Histogram", stat="density")

# Plot PDF using KDE (Kernel Density Estimation)
sns.kdeplot(data["GraduationYear"], color="blue", label="PDF")

plt.title("Distribution of Graduation Year")
plt.legend()
plt.show()
```

### **Observations**
#### ======> This PDF and histogram plot reveals that the majority of employees are completed their GraduationYear in the year 2013.

```python
# Histogram and PDFs of CollegeTier
plt.figure(figsize=(8, 5))
plt.xticks(rotation=90)

# Plot Histogram
sns.histplot(data["CollegeTier"], color="red", kde=False, label="Histogram", stat="density")

# Plot PDF using KDE (Kernel Density Estimation)
sns.kdeplot(data["CollegeTier"], color="blue", label="PDF")

plt.title("Distribution of CollegeTier")
plt.legend()
plt.show()
```

### **Observations**
#### =====> The diagram shows that the most of employees are concentrated in the highest CollegeTier category (2.0), indicating a significant skew towards that tier.

```python
# Histogram and PDFs of Salary Column
plt.figure(figsize=(8, 5))
plt.xticks(rotation=90)

# Plot Histogram
sns.histplot(data["Salary"], color="red", kde=False, label="Histogram", stat="density")

# Plot PDF using KDE (Kernel Density Estimation)
sns.kdeplot(data["Salary"], color="blue", label="PDF")

plt.title("Distribution of Salary")
plt.legend()
plt.show()
```

### **Observations**
#### =====> The diagram shows that most employees have salaries concentrated at the lower end of the range, with a sharp peak indicating a high density of lower salaries. The density decreases as the salary increases.

### **kdeplot in all educational percentages**

- #### **A Kernel Density Estimation (KDE) plot is a non-parametric way to estimate the probability density function of a continuous random variable.**

- #### **KDE plot can be used to visualize the distribution of a single continuous variable. It is similar to a histogram but provides a smoother curve instead of discrete bins.**

```python
# KDE plot of 10percentage
plt.figure(figsize=(8, 5))
sns.kdeplot(data["10percentage"], color="k")

# Add title and labels
plt.title("Kernel Density Estimation of 10th Percentage")
plt.xlabel("10th Percentage")
plt.ylabel("Density")

# Show the plot
plt.show()
```

### **Observations**
#### =====> This KDE plot illustrates the density of employees' 10th percentage, revealing a left-skewed distribution where the majority of employees have around 85% in their 10th-grade exams.

```python
# KDE plot of 12ercentage
plt.figure(figsize=(8, 5))
sns.kdeplot(data["12percentage"], color="k")

# Add title and labels
plt.title("Kernel Density Estimation of 12th Percentage")
plt.xlabel("12th Percentage")
plt.ylabel("Density")

# Show the plot
plt.show()
```

### **Observations**
#### =====> This KDE plot shows the density distribution of employees' 12th-grade percentages, resembling a normal distribution with the majority falling between 70% to 80%, indicating a common range for most employees' 12th-grade scores.

```python
# KDE plot of collegeGPA
plt.figure(figsize=(8, 5))
sns.kdeplot(data["collegeGPA"], color="k")

# Add title and labels
plt.title("Kernel Density Estimation of collegeGPA")
plt.xlabel("collegeGPA")
plt.ylabel("Density")

# Show the plot
plt.show()
```

### **Observations**
#### =====> This KDE plot illustrates the density distribution of employees' college GPA, resembling a normal distribution with the majority falling between 65 to 75, suggesting a common range for most employees' college GPAs.

## **Step - 5 - Research Questions**

- #### **Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.**

- #### **Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?**


```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t, norm
import statistics
import random

# Filter the DataFrame for relevant designations
data_research = data[(data["Designation"]=="programmer analyst")|(data["Designation"]=="software engineer")|(data["Designation"]=="hardware engineer")
       |(data["Designation"]=="associate engineer")]
data_research

# OR
# data_research = data[(data["Designation"].isin(["programmer analyst", "software engineer", "hardware engineer", "associate engineer"]))]
```

```python
# Extract the salary data from the data_research
salary_data = data_research["Salary"]
salary_data
```

```python
# Convert the salary data to a list
salary_list = []
for salary in salary_data:
    salary_list.append(salary)
print(salary_list)
```

```python
# Generate a random sample of salary data
num = 50

# Randomly select 50 salaries from "salary_list"
sample = random.sample(salary_list, num)
print(sample)
```

```python
# Calculate sample statistics: mean and standard deviation
sample_mean = np.mean(sample)
sample_std = statistics.stdev(sample)

# Assuming population mean is 300000
pop_mean = 300000  

# Calculate the sample size
sample_size = len(sample)
```

```python
# Output the sample statistics
print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"Sample Size: {sample_size}")
print(f"Sample population mean: {pop_mean}")
```

```python
# Function to calculate t-score
def t_score(sample_size, sample_mean, pop_mean, sample_std):
    numerator = sample_mean - pop_mean
    denominator = sample_std / (sample_size ** 0.5)
    return numerator / denominator

# Calculate the t-value for the given sample
t_value = t_score(sample_size, sample_mean, pop_mean, sample_std)
print(f"T-Value: {t_value}")

# Set confidence level and calculate alpha
confidence_level = 0.95
alpha = 1 - confidence_level
print(f"Alpha: {alpha}")

# Calculate the critical t-value using degrees of freedom (sample_size - 1)
t_critical = t.ppf(1 - alpha / 2, df=sample_size - 1)
print(f"T-Critical: {t_critical}")
```

```python
# Plot the normal distribution and the critical regions
x_min = -200000  
x_max = 800000   
mean = pop_mean  
std = sample_std 

# Generate x values for plotting the normal distribution curve
x = np.linspace(x_min, x_max, 100)

# Calculate the y values for the normal distribution PDF
y = norm.pdf(x, mean, std)

# Plot the normal distribution curve
plt.xlim(x_min, x_max)
plt.plot(x, y, label='Normal Distribution')

# Calculate the critical values for the left and right tails
t_critical_left = mean + (-t_critical * std)
t_critical_right = mean + (t_critical * std)

# Plot and fill the left critical region
x1 = np.linspace(x_min, t_critical_left, 100)
y1 = norm.pdf(x1, mean, std)
plt.fill_between(x1, y1, color='blue', alpha=0.5, label='Critical Region')

# Plot and fill the right critical region
x2 = np.linspace(t_critical_right, x_max, 100)
y2 = norm.pdf(x2, mean, std)
plt.fill_between(x2, y2, color='blue', alpha=0.5)

# Mark and label the sample mean on the plot
plt.scatter(sample_mean, 0, color='red', label='Sample Mean')
plt.annotate("Sample Mean", (sample_mean, 0.7), color='red')

# Add title and labels
plt.title("T-Test: Critical Regions and Sample Mean")
plt.xlabel("Salary")
plt.ylabel("Probability Density Function (PDF)")
plt.legend()
plt.show()
```

```python
# Compare t-value with critical t-value
if np.abs(t_value) > t_critical:
    print("Reject Null Hypothesis based on t-value.")
else:
    print("Do not reject the Null Hypothesis based on t-value.")
```

```python
# Calculate the p-value (2-tailed test)
p_value = 2 * (1.0 - t.cdf(np.abs(t_value), df=sample_size - 1))
print("p-value =", p_value)

# Compare p-value with alpha (significance level)
if p_value < alpha:
    print("Reject Null Hypothesis based on p-value.")
else:
    print("Do not reject the Null Hypothesis based on p-value.")
```
